1. What is a parameter?


A parameter is like a setting inside a model that help it learn.

It changes while training to get better results.

It's kinda like a dial you turn to get best output.

Not same as input data, it's more like hidden part.

Model uses them to learn patterns.


2. What is correlation? What does negative correlation mean?


Correlation means how two things move together.

* If one go up and other also go up, it is positive.

* If one go up and other go down, it is negative.

* No pattern means no correlation.

* It shows relation, but not cause.

Negative Correlation means when one thing increases, other thing decrease.

* Like if you study more, your mistakes go down.

* Both move in opposite direction.

* It's not always 100% sure, just a trend.

* It help us understand connection between things.

4. How does loss value help in determining whether the model is good or not?


Loss tells how wrong the model is.

Small loss means model is doing good.

Big loss means it making more mistakes.

Model try to reduce loss during training.

It's like a score for how bad the guess is.


5. What are continuous and categorical variables?


Continuous variables are numbers that go on and on, like height or age.

They can have lots of values, even in between like 5.6 or 7.3.

Categorical variables are like labels, such as color or country.

They are not numbers, just groups or types.

Model use them different ways.


6. How do we handle categorical variables in Machine Learning? What are the common techniques?

We can't use text directly, so we change it to numbers.

One common way is One-Hot Encoding, makes new columns for each value.

Another way is Label Encoding, just give numbers to each label.

Sometimes we also use Target Encoding.

This help model to understand non-number stuff.

21. What is feature scaling? How does it help in Machine Learning?

Feature scaling makes all numbers in data same range.

Like changing big numbers to small range like 0 to 1.

It helps model to learn better and faster.

Without scaling, model get confused by big values.

Important for things like distance-based models.

22. How do we perform scaling in Python?

We use StandardScaler or MinMaxScaler from sklearn.

StandardScaler makes data like normal bell curve.

MinMaxScaler puts values between 0 and 1.

We fit the scaler on data, then transform it.

Scaling is before training the model.

25. Explain data encoding?

Encoding means changing text data to numbers.

Model don't understand words, only numbers.

Like turning “red” to [1, 0, 0] and “blue” to [0, 1, 0].

It help model work with category data.
